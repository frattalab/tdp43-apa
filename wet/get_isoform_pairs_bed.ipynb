```python
import pyranges as pr
import pandas as pd
from helpers import get_terminal_regions
import os
```

```python
pairs_df = pd.read_csv("data/regions/seddighi_i3_cryptic_annot_pairs.csv", index_col=False)
pairs_df
```

```python
d3utr = pr.read_bed("data/regions/2023-12-15_papa_cryptic_d3utr.background_shsy5y.last_exons.bed")
d3utr
```

```python
if not os.path.exists("data/regions/gencode.v40.annotation.TLX1.gtf"):
    os.system("grep 'TLX1' data/regions/gencode.v40.annotation.gtf > data/regions/gencode.v40.annotation.TLX1.gtf")
```

```python
# extract exons of TLX1 transcripts from GTF
tlx1_gtf = pr.read_gtf("data/regions/gencode.v40.annotation.TLX1.gtf")
tlx1_gtf =tlx1_gtf.subset(lambda df: df.Feature == "exon")
tlx1_gtf.exon_number = tlx1_gtf.exon_number.astype(int)
tlx1_gtf
```

```python
# get last exons for TLX1 transripts
tlx1_les = get_terminal_regions(tlx1_gtf)
tlx1_les
```

```python
pairs_df
```

```python
# Create 3'UTR extension BED
# first - make sure select TLX1 event from annotation, as proximal PAS was mis-annotated

# for TLX1, extract 'ENST00000370196.11' from refernce GTF (proximal is currently mis-annotated)
# get tx id
tlx1_prox_tx = pairs_df.loc[(pairs_df.gene_name == "TLX1") & ~(pairs_df.bed_name.str.contains("distal", regex=False)), "bed_name"].iloc[0]

# extract last exon for selected transcript, 
tlx1_annot_le = (tlx1_les.subset(lambda df: df.transcript_id == tlx1_prox_tx)                            
 .assign("annot_status", lambda df: pd.Series(["annotated"]*len(df), index=df.index))
 .assign("Score", lambda df: pd.Series(["."]*len(df), index=df.index))
 .assign("Name", lambda df: df.gene_name.str.cat(df[["annot_status", "transcript_id"]].astype(str), sep="|"))
 [["Name", "Score"]]
.sort()
)

tlx1_annot_le
#tlx1_gtf
```

```python
#1. Select proximal and distal pairs for SIX3 and ELK1
d3utr_region_nms = pairs_df.loc[pairs_df.gene_name.isin(["ELK1", "SIX3", "TLX1"]), "bed_name"].to_list()
d3utr_sel = d3utr.subset(lambda df: df.Name.isin(d3utr_region_nms))
d3utr_sel = pr.concat([d3utr_sel, tlx1_annot_le]).sort()
d3utr_sel
```

```python
d3utr_sel.to_bed("processed/validation_region_beds/2024-04-08_cryptic_and_annotated.selected.3ext.bed")
```

```python
## IPA events
ipa = pr.read_bed("processed/validation_region_beds/2024-04-05_cryptic_and_annotated.ipa.bed")
ipa_region_gns = pairs_df.loc[pairs_df.gene_name.isin(["SIN3B", "ANKRD27", "CNPY3"]), "bed_name"].to_list()
ipa_sel = ipa.subset(lambda df: df.Name.isin(ipa_region_gns))
ipa_sel
```

```python
import numpy as np
# IPA - Update SIN3B's PAS - slight extension of ~ 200 bp (by coverage and jnc read support)
# name: chr19:+:16857609:16857610 location: chr19:16857607-16857612 (+) score: 18.0
# name is in BED format, so just update end of SIN3B cryptic to 16857610
ipa_sel = ipa_sel.assign("End",
               lambda df: pd.Series(np.where(df["Name"] == "ENSG00000127511.10_2|SIN3B|bleedthrough|cryptic", 16857610, df["End"])))
ipa_sel
```

```python
ipa_sel.sort().to_bed("processed/validation_region_beds/2024-05-13_cryptic_and_annotated.selected.ipa.bed")
```

```python
# ALEs
ale = pr.read_bed("processed/validation_region_beds/2024-04-05_cryptic_and_annotated.ales.bed")
ale_region_gns = pairs_df.loc[pairs_df.gene_name.isin(["STMN2", "ARHGAP32", "PHF2", "SYNJ2"]), "bed_name"].to_list()
ale_sel = ale.subset(lambda df: df.Name.isin(ale_region_gns))
ale_sel
```

```python
# ARHGAP32
# ARHGAP32 has two PAS, one major (in current BED + supported by 99 junction reads (name: chr11:-:128991449:128991450,  location: chr11:128991448-128991461 (-), # score: 99.0)
# and 1 more distal, minor PAS found in i3Ns which closely aligns with PolyASite site
# name: 11:128988275:- , location: chr11:128988275-128988275 (-) score: 0.0011
# 
# Need to add minor PAS
arhgap32_extra = ale_sel[ale_sel.Name == "ENSG00000134909.19_1|ARHGAP32|spliced|cryptic"]
arhgap32_extra.Start = 128988275 -1 # PolyASite uses UCSC position format, so subtract 1 to get BED-format coord
arhgap32_extra.Name = "ENSG00000134909.19_1;11:128988275:-|ARHGAP32|spliced|cryptic"
arhgap32_extra

ale_sel = pr.concat([ale_sel, arhgap32_extra])
ale_sel
```

```python
# STMN2
# 1. Current representative event in BED (+ assembled) - name: chr8:+:79617076:79617077 location: chr8:79617065-79617080 (+) score: 224.0
# 2. Additional slightly lower supported PAS approx 130bp downstream - name: chr8:+:79617206:79617207 location: chr8:79617199-79617213 (+) score: 143.0
stmn2_extra = ale_sel[ale_sel.Name == "ENSG00000104435.14_1|STMN2|spliced|cryptic"]
stmn2_extra.End = 79617207
stmn2_extra.Name = "ENSG00000104435.14_1;chr8:+:79617206:79617207|STMN2|spliced|cryptic"
ale_sel = pr.concat([ale_sel, stmn2_extra])
ale_sel
```

```python
# PHF2
# 1. 1 site supported by junction reads ~ 33bp upstream - name: chr9:+:93661958:93661959 location: chr9:93661957-93661967 (+) score: 5.0
# 2. 1 junction read site ~ 150 bp upstream - name: chr9:+:93661875:93661876 location: chr9:93661865-93661876 (+) score: 9.0
phf2_extra1 = ale_sel[ale_sel.Name == "ENSG00000197724.11_1|PHF2|spliced|cryptic"]
phf2_extra2 = phf2_extra1.copy()
phf2_extra1.End = 93661959
phf2_extra1.Name = "ENSG00000197724.11_1;chr9:+:93661958:93661959|PHF2|spliced|cryptic"
phf2_extra2.End = 93661876
phf2_extra2.Name = "ENSG00000197724.11_1;chr9:+:93661875:93661876|PHF2|spliced|cryptic"
phf2_extras = pr.concat([phf2_extra1, phf2_extra2])
phf2_extras
```

```python
ale_sel = pr.concat([ale_sel, phf2_extras]).subset(lambda df: ~(df.Name == "ENSG00000197724.11_1|PHF2|spliced|cryptic")).sort()
ale_sel
```

```python
ale_sel.to_bed("processed/validation_region_beds/2024-05-13_cryptic_and_annotated.selected.ale.bed")
```
